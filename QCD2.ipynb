```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
tm=15
Q1 = np.loadtxt('Q_12_280.out')
Q1square=np.mean(Q1)**2
Gamma1=np.zeros(tm) 
for i in range(tm):
    for j in range(0,len(Q1)-tm):
        Gamma1[i]+=Q1[j]*Q1[i+j]
    Gamma1[i]/=(len(Q1)-tm)
    Gamma1[i]-=Q1square

Q2 = np.loadtxt('1_16b_400.out')
Q2square=np.mean(Q2)**2
Gamma2=np.zeros(tm) 
for i in range(tm):
    for j in range(0,len(Q2)-tm):
        Gamma2[i]+=Q2[j]*Q2[i+j]
    Gamma2[i]/=(len(Q2)-tm)
    Gamma2[i]-=Q1square

Q3 = np.loadtxt('1_16_480.out')
Q3square=np.mean(Q3)**2
Gamma3=np.zeros(tm) 
for i in range(tm):
    for j in range(0,len(Q3)-tm):
        Gamma3[i]+=Q3[j]*Q3[i+j]
    Gamma3[i]/=(len(Q3)-tm)
    Gamma3[i]-=Q3square

plt.plot(Gamma1/Gamma1[0], 'r', label='Reticolo B1')
plt.plot(Gamma2/Gamma2[0], 'b',  label='Reticolo B2')
plt.plot(Gamma3/Gamma3[0], 'g',  label='Reticolo B3')
#plt.xlim(0-0.01, 0.4)
#plt.ylim(0.00045,0.00075)
plt.xlabel(r'$t_M$')
plt.ylabel(r'$\Gamma(t_M)$')
plt.title('Funzione di autocorrelazione ')
plt.legend()
plt.show()

```
